```python
#importando
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
from matplotlib import pyplot as plt
import plotly.express as px

# Procurando ao da Via Varejo
asset_viarejo = 'VIIA3.SA'

# DATAS
start = '2019-01-01'
end = "2023-07-24"

# PREOS
prices = pdr.get_data_yahoo(asset_viarejo, start=start, end=end)['Adj Close']

# CRIANDO GRFICO
fig = px.line(prices.reset_index(), x='Date', y='Adj Close', title='Grfico da Ao da Via Varejo')
fig.show()
```
