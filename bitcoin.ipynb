```python

%matplotlib inline

import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime

# Download data from yfinance
bitcoin = yf.Ticker("BTC-USD")
data = bitcoin.history(start="2010-01-01", end=datetime.now().strftime('%Y-%m-%d'))

# Create the plot
fig = plt.figure(figsize=(10, 6), dpi=100, facecolor="white")
ax = fig.add_subplot(111)
ax.plot(data['Close'], color='blue', label="Preço de Fechamento Diário")
ax.legend()
plt.title("Histórico de Preços do Bitcoin de 2015 a 2023", fontsize=18)
plt.xlabel("Data", fontsize=14)
plt.ylabel("Preço (USD)", fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.grid(axis="y")

# Add interactivity
def on_move(event):
    if not event.inaxes:
        return
    x, y = event.xdata, event.ydata
    date = datetime.fromtimestamp(x).strftime('%Y-%m-%d')
    plt.title(f"Preço do Bitcoin: {y:.2f} USD em {date}", fontsize=16)

fig.canvas.mpl_connect("motion_notify_event", on_move)
plt.text(data.index[0], -15000, "Data source: Yahoo Finance", fontsize=10)
plt.tight_layout()
plt.savefig('bitcoin1_price.png')
plt.show()
```
