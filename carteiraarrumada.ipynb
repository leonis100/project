```python
#importando
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
from matplotlib import pyplot as plt 
import plotly.express as px
```

```python
#PROCURANDO A AÇÃO
assets = [ 'VIIA3.SA', 'MSFT34.SA', 'PFIZ34.SA','NVDC34.SA','GOGL34.SA','BOAC34.SA','AMZO34.SA','JPMC34.SA','AAPL34.SA','MGLU3.SA']
```

```python
#DATAS
start = '2020-01-01'
end= "2022-10-08"
```

```python
#PREÇOS
prices=pdr.get_data_yahoo (assets, start= start ) [ 'Adj Close']
```

```python
#PARAMETROS
returns = prices.pct_change().dropna()
weights = np.array([0.10, 0.10, 0.10, 0.10,0.10,0.10,0.10,0.10,0.10,0.10])
return1 = returns.assign(portfolio = returns.dot(weights))
```

```python
#CRIANDO GRÁFICO E EXCEL
acum_index = (1+return1).cumprod()
cum = acum_index.reset_index(level = 0)
plt.figure(figsize = (20, 20))
fig = px.line(cum, x = 'Date', y = assets + ['portfolio'])
fig.show()
cum.head
cum.to_excel('ações2.xls')
prices.to_excel('val.xlsx')
```

```python
#GRÁFICO PREÇOS
acum= prices.reset_index(level=0)
fig=px.line(acum,x="Date",y=assets)
fig.show()
```
